```python
budget = {
    "Food": 0,
    "Transport": 0,
    "Education": 0,
    "Entertainment": 0
}

def display_report(budget):
    print("\n--- Final Expense Report ---")
    total_spent = 0
    highest_category = None
    highest_amount = -1

    for category, amount in budget.items():
        print(f"{category}: ${amount:.2f}")
        total_spent += amount
        if amount > highest_amount:
            highest_amount = amount
            highest_category = category

    print(f"\nTotal spent across everything: ${total_spent:.2f}")
    if highest_category:
        print(f"Highest spending category: {highest_category} (${highest_amount:.2f})")
    else:
        print("No expenses recorded.")

while True:
    category_input = input("What category did you spend on today? (e.g., Food, Transport, Education, Entertainment) Type 'report' for summary or 'exit' to quit: ").strip().capitalize()

    if category_input == 'Exit':
        break
    elif category_input == 'Report':
        display_report(budget)
        continue


    if category_input in budget:
        while True:
            try:
                amount_input = float(input(f"How much did you spend on {category_input}? "))
                if amount_input < 0:
                    print("Amount cannot be negative. Please enter a positive number.")
                else:
                    budget[category_input] += amount_input
                    print(f"Added ${amount_input:.2f} to {category_input}.")
                    break
            except ValueError:
                print("Invalid amount. Please enter a number.")

    else:
        print("Category not found. Please choose from Food, Transport, Education, Entertainment, or type 'exit' or 'report'.")

display_report(budget)
```
