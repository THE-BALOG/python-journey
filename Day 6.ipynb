```python
def main():
  while True:
    user_input = input("Do you want to run the report? (yes/no): ")
    if user_input == "yes":
        student_report()
        break
    elif user_input == "no":
        print("ok")
        break
    else:
        print("input either yes or no")

def student_report():
  students = [
    {"name" : "David", "score" : 90,},
    {"name" : "Daniel", "score" :78,},
    {"name" : "Dominic", "score" :89,},
    {"name" : "Dotun", "score" :90,},
    {"name" : "Drake", "score" :45}
  ]

  for student in students:
    score = student["score"]
    if score < 50:
      print(f"Skipping {student['name']} - score too low")
      continue

    if score >= 70 :
      print(student["name"], score, "Grade A", sep=" - ")
    elif 50 <= score <= 69:
      print(student["name"], score, "Grade B", sep=" - ")
    else:
      print(student["name"], score, "Grade C", sep=" - ")

main()
```
