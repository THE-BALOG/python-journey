```python
import random
import statistics
```

```python
random_numbers = [random.randint(1, 100)
for _ in range(10)]
print(random_numbers)
```

```python
mean_value = statistics.mean(random_numbers)
median_value = statistics.median(random_numbers)
mode_value = statistics.multimode(random_numbers)

print("The average (mean) number is:", mean_value)
print("The middle (median) number is:", median_value)
print("The most popular (mode) number is:", mode_value)
```

```python
highest_number = max(random_numbers)
lowest_number = min(random_numbers)

print("The highest number is:", highest_number)
print("The lowest number is:", lowest_number)
```
