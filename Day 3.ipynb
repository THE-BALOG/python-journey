```python
def check_age():
  age = int(input("What is your age ?"))
  if age >= 18:
    print("You are an adult.")
  elif 13 <= age <= 17:
    print("You are a teenager")
  else:
    print("You are a child")

check_age()

```

```python
def grade_checker():
  name = input("What is your name? ").strip().title()
  score = int(input("what is your Score ? "))
  if score >= 90:
    print(f"Excellent {name}! You got an A")
    print("Passed: True")
  elif 80 <= score < 90:
    print(f"Great job {name}! You got a B")
    print("Passed: True")
  elif 70 <=  score < 80:
    print(f"Good effort {name}! You got a C")
    print("Passed: True")
  elif 60 <=  score < 70:
    print(f"You passed {name}! You got a D")
    print("Passed: True")
  else:
    print(f"You failed {name}. Keep pushing.")
    print("Passed: False")

grade_checker()
```
